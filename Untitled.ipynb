```python
import pandas as pd

df = pd.read_csv('pokemon_data.csv')

print(df.head(5))
```

```python
#read headers
print(df.columns)

#read each column
print(df['Name'][0:5]) #or df.Name

print(df[['Name', 'Type 1']][0:5])

#read each row
print(df.iloc[0:4]) #integer location


print(df.loc[df['Type 1'] == 'Fire'])#find specific data that isnt int based - textual info

#read specific location (R,C)
print(df.iloc[2,1]) #row then col
```

```python
#describing data

df.describe() #summary information
```

```python
#sorting

df.sort_values(['Type 1', 'HP'], ascending = [1, 0])
```

```python
#changes to the data

#create a total category
df['Total'] = df.iloc[:, 4:10].sum(axis = 1) #axis 1 horizontal

cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df.head(5)
```

```python
#filter data
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]

#reset index
new_df.reset_index(drop = True, inplace = True)

new_df
```

```python
import re

#remove rows with Mega in name
#df.loc[~df['Type 1'].str.contains('Fire|Grass', flags = re.I, regex = True)] 
#~ means not 
# re.I means ignore capitals

df.loc[df['Name'].str.contains('^pi[a-z]*', flags = re.I, regex = True)] 
#starts with PI

```

```python
#conditional changes
df.loc[df['Type 1'] == 'Flamer', 'TLenegdary'] = True

df
```

```python
#aggregate statistics
#group by mean and sort values can also do sum(), count()
df.groupby(['Type 1']).mean().sort_values('Defense', ascending = False)
```

```python
df['count'] = 1

df.groupby(['Type 1', 'Type 2']).count()['count']
```

```python
#large dataset loading
new_df = pd.DataFrame(columns = df.columns)

for df in pd.read_csv('pokemon_data.csv', chunksize = 5):
    results = df.groupby(['Type 1']).count()
    
    new_df = pd.concat([new_df, results])
#loads DF but in chunks of 5

new_df
```
